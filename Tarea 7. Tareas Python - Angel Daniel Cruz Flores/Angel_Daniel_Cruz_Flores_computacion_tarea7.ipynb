### Ejercicio 1 (1 punto):
En una celda Markdown contesta: ¿Qué es un nombre reservado en python? Enlista 5 ejemplos de nombres reservados en python y describe su función (aquí debes investigar su función si es que aún no hemos tratado dicho nombre reservado).

Python tiene un conjunto de palabras reservadas que no podemos utilizar para nombrar variables ni funciones, ya que las reserva internamente para su funcionamiento.

Es una palabra en inglés a la que no se le puede asignar un valor númerico, ni aplicarle otros comandos relacionados con ésto, pues éstos, precisamente, ya están reservados para funciones fundamentales para el lenguge de programación de Python.

Aquí unos ejemplos de nombres reservados, junto con su función:

**if** = En Python, la sentencia if se utiliza para ejecutar un bloque de código si, y solo si, se cumple una determinada condición. Por tanto, if es usado para la toma de decisiones. Es decir, solo si condición se evalúa a True, se ejecutarán las sentencias que forman parte de bloque de código. En caso de que se evalúe a False no se ejecutará ninguna sentencia perteneciente a bloque de código.

**elif** = También es posible que te encuentres situaciones en que una decisión dependa de más de una condición.
En estos casos se usa una sentencia if compuesta. Es decir, en esta ocasión, se comprueba la condición cond1. Si se evalúa a True, se ejecuta el bloque bloque cond1 y después el flujo continúa después del bloque if.
Sin embargo, si cond1 se evalúa a False, se comprueba la condición cond2. Si esta se evalúa a True, se ejecuta el bloque de sentencias bloque cond2. En caso contrario, pasaría a comprobar la condición del siguiente elif y así sucesivamente hasta que llegue al else, que se ejecuta si ninguna de las condiciones anteriores se evaluaron a True.

**else** = Hay ocasiones en que la sentencia if básica no es suficiente y es necesario ejecutar un conjunto de instrucciones o sentencias cuando la condición se evalúa a False.
Para ello se utiliza la estructura if ... else...

**while** = El bucle while es otra estructura de control de flujo, concretamente lo que hace es repetir un código mientras dure una determinada condición. 

**print** = Imprime un valor, oración, lista, etc., especificado en los paréntesis. 

**for** = El bucle for se utiliza para recorrer los elementos de un objeto iterable (lista, tupla, conjunto, diccionario, …) y ejecutar un bloque de código. En cada paso de la iteración se tiene en cuenta a un único elemento del objeto iterable, sobre el cuál se pueden aplicar una serie de operaciones.

**True, False** = Una variable booleana es una variable que sólo puede tomar dos posibles valores: True (verdadero) o False (falso). 

**def** = def es una palabra reservada que indica a Python que una nueva función está siendo definida.

*   **Los pasos para encontrar sus raíces a través del método de la chicharronera.**

  Sea $f(x) = 10x^{2} - 2x$ \\

  Entonces: \\
  
      $10x^{2} - 2x = 0$ \\

*     Simplificamos la expresión dividiendo entre dos: 

    $5x^{2} - x = 0$ \\

    $5x^{2} - x - 0= 0$

    $a = 5$, $b = -1$ y $c = 0$ 

*   Usando la ecuación cuadrática (chicharronera) tenemos: \\

    $x_{1,2} = \frac {-(-1) \pm \sqrt{(-1)^{2} - 4 (5)(0)} }{2(5)}$ \\

    $x_{1,2} = \frac {1 \pm \sqrt{1-0} }{10}$ \\

    $x_{1,2} = \frac {1 \pm 1 }{10}$ \\

*   Resolvemos las dos posibles soluciones quedandonos: \\

    $x_{1} = \frac {1 + 1 }{10} = \frac {2}{10} = \frac {1}{5}$  \\
    $x_{2} = \frac {1 - 1 }{10} = \frac {0}{10} = 0 $

    $x_{1} = \frac{1}{5}$  \\
    $x_{2} =  0 $

```python
#Importamos las librerias necesarias
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px

#Hacemos que eliga un valor aleatorio entre dos valores: X y Y y repetimos el volado 100000 veces
volados = np.random.randint(low=0, high=2, size=100000)
#Hacemos que nos de los resultados en una tabla
tabla_volados = pd.DataFrame(volados, columns=["Resultados"])
#Remplazamos los valores X y Y por los posibles resultados: Águila y Sol
tabla_volados.Resultados.replace({0:"Aguila", 1:"Sol"}, inplace=True)

#Finalmente, graficamos el histograma
fig = px.histogram(tabla_volados, text_auto=True ,labels={
                      "value": "Resultado",
                      "count": "Frecuencia"}, template='plotly_white')
fig.update_layout(
    title="Simulación de tirar el volado 100000 veces ",
    xaxis_title="Frecuencia",
    yaxis_title="Resultado"
)
```
